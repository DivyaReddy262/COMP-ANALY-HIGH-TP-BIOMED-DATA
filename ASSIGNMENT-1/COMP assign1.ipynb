```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display  

# Load matrices from files
matrix1_path = r"C:\Users\kdivy\OneDrive\Desktop\COMP ANA HT DATA\matrix1.txt"
matrix2_path = r"C:\Users\kdivy\OneDrive\Desktop\COMP ANA HT DATA\matrix2.txt"

# Read data 
matrix1 = pd.read_csv(matrix1_path, sep="\t", index_col=0)
matrix2 = pd.read_csv(matrix2_path, sep="\t", index_col=0)

# Drop "Unnamed" column if it exists (NaN values)
matrix1 = matrix1.drop(columns=[col for col in matrix1.columns if "Unnamed" in col], errors='ignore')
matrix2 = matrix2.drop(columns=[col for col in matrix2.columns if "Unnamed" in col], errors='ignore')

# Extract only the **first 12 cancer types** to form a 12×12 table
matrix1_12x12 = matrix1.iloc[:, :12]
matrix2_12x12 = matrix2.iloc[:, :12]

# Compute Pearson correlation matrices (12×12)
corr_matrix1 = matrix1_12x12.corr(method="pearson")
corr_matrix2 = matrix2_12x12.corr(method="pearson")

# Display correlation matrices
print("Correlation Matrix 1:")
display(corr_matrix1)
matrix1_correlation = corr_matrix1[["UCEC"]]
# Display the extracted correlation column properly
display(matrix1_correlation)

print("\nCorrelation Matrix 2:")
display(corr_matrix2)
matrix2_correlation = corr_matrix2[["UCEC"]]
# Display the extracted correlation column properly
display(matrix2_correlation)


# Generate heatmaps for both matrices
sns.heatmap(corr_matrix1, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Pearson Correlation - Matrix 1")
plt.show()
sns.heatmap(corr_matrix2, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Pearson Correlation - Matrix 2")
plt.show()

```
